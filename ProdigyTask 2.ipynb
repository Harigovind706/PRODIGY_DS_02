The Titianic dataset contains the data of the people who were in the titanic and further variables include survival, Pclass, Sex, count of family members of each passenger with unique id, ticket, Fare, Cabin and Embarked. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
titanic_data = pd.read_csv("C:/Users/Hari/Desktop/Prodigy Infotech/titanic.csv")

# Check the first few rows
print(titanic_data.head())

# Check the shape of the dataset
print(titanic_data.shape)

# Check the data types of columns
print(titanic_data.info())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())

# Fill missing values in 'Age' with median age
median_age = titanic_data['Age'].median()
titanic_data['Age'].fillna(median_age, inplace=True)

# For 'Cabin', as it has many missing values, we might consider dropping it for now
titanic_data.drop('Cabin', axis=1, inplace=True)
```

```python
# Numerical features
titanic_data[['Age', 'Fare']].describe()

# Categorical features
print(titanic_data['Survived'].value_counts())
print(titanic_data['Pclass'].value_counts())
print(titanic_data['Sex'].value_counts())
print(titanic_data['Embarked'].value_counts())

# Visualizations
sns.countplot(x='Survived', data=titanic_data)
sns.countplot(x='Pclass', data=titanic_data)
sns.countplot(x='Sex', data=titanic_data)
sns.countplot(x='Embarked', data=titanic_data)
```

```python
# Relationship between survival and categorical features
sns.countplot(x='Survived', hue='Sex', data=titanic_data)
```

```python
sns.countplot(x='Survived', hue='Embarked', data=titanic_data)
```

```python
titanic_data['Pclass_category'] = titanic_data['Pclass'].astype(str)
sns.countplot(x='Survived', hue='Pclass_category', data=titanic_data)
```

```python
# Relationship between age and Pclass features
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=titanic_data, width=0.5)
plt.grid(True)
plt.title('Age Distribution by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.show()
```
